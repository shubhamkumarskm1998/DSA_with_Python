# Linked List in Python

## What is a Linked List?

A **Linked List** is a linear data structure where elements (nodes) are connected using references (pointers).

Each node contains:

* Data
* Reference to the next node

```text
10 → 20 → 30 → 40 → None
```

Unlike arrays, linked list elements are not stored in contiguous memory locations.

---

## Node Structure

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

Example:

```python
n1 = Node(10)
n2 = Node(20)

n1.next = n2
```

```text
10 → 20 → None
```

---

## Creating a Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
```

```text
10 → 20 → 30 → None
```

---

## Traversing a Linked List

```python
current = head

while current:
    print(current.data)
    current = current.next
```

Output:

```text
10
20
30
```

---

## Insertion Operations

### Insert at Beginning

```python
new_node = Node(5)
new_node.next = head
head = new_node
```

```text
5 → 10 → 20 → 30
```

### Insert at End

```python
new_node = Node(40)

current = head
while current.next:
    current = current.next

current.next = new_node
```

```text
10 → 20 → 30 → 40
```

### Insert After a Node

```python
new_node = Node(25)

new_node.next = current.next
current.next = new_node
```

```text
10 → 20 → 25 → 30
```

---

## Deletion Operations

### Delete First Node

```python
head = head.next
```

### Delete Last Node

```python
current = head

while current.next.next:
    current = current.next

current.next = None
```

### Delete Specific Value

```python
key = 20

current = head

while current.next:
    if current.next.data == key:
        current.next = current.next.next
        break
    current = current.next
```

---

## Searching

```python
key = 30

current = head

while current:
    if current.data == key:
        print("Found")
        break
    current = current.next
```

---

## Reverse a Linked List

```python
prev = None
current = head

while current:
    nxt = current.next
    current.next = prev
    prev = current
    current = nxt

head = prev
```

Before:

```text
10 → 20 → 30 → None
```

After:

```text
30 → 20 → 10 → None
```

---

## Types of Linked Lists

### 1. Singly Linked List

```text
10 → 20 → 30 → None
```

### 2. Doubly Linked List

```text
None ← 10 ⇄ 20 ⇄ 30 → None
```

Each node has:

* prev pointer
* next pointer

### 3. Circular Linked List

```text
10 → 20 → 30
↑         ↓
└─────────┘
```

Last node points back to the first node.

---

## Time Complexity

| Operation           | Complexity |
| ------------------- | ---------- |
| Access              | O(n)       |
| Search              | O(n)       |
| Insert at Beginning | O(1)       |
| Insert at End       | O(n)       |
| Delete Beginning    | O(1)       |
| Delete End          | O(n)       |

---

## Array vs Linked List

| Array                   | Linked List           |
| ----------------------- | --------------------- |
| Contiguous memory       | Non-contiguous memory |
| Fast access O(1)        | Slow access O(n)      |
| Fixed/Resizing overhead | Dynamic size          |
| Insert/Delete costly    | Insert/Delete easier  |

---

## Interview One-Liner

> A Linked List is a dynamic linear data structure consisting of nodes, where each node stores data and a reference to the next node, allowing efficient insertion and deletion without requiring contiguous memory.

````python
### Class singlylinkedlist

### Delete Operations

class singlylinkedlist:
    def __init__(self,head=None):
        self.head = head

    def insertatend(self,value):
        temp = Node(value)
        if self.head !=None:
            t1 = self.head
            while t1.next != None:
                t1 = t1.next
            t1.next = temp
        else:
            self.head = temp

    def printlinklist(self):
        t1 = self.head
        while t1.next!=None:
            print(t1.data,t1.next)
            t1 = t1.next
        print(t1.data,t1.next)

    def insertatbeg(self,value):
        temp = Node(value)
        temp.next = self.head
        self.head = temp

    def insertatnode(self,value,x):
        temp = Node(value)
        t1= self.head
        while t1.next != None:
            if t1.data == x:
                temp.next = t1.next
                t1.next = temp
            t1 = t1.next

    def delete(self,value):
        t1= self.head
        prev = t1
        if t1.data == value:
            self.head = t1.next
        while t1 != None:
            if t1.data == value:
                prev.next = t1.next
                break
            else:
                prev = t1
                t1 = t1.next
        if t1.data == value:
            prev.next = None

obj = singlylinkedlist()
obj.insertatend(5)
obj.insertatend(10)
obj.insertatend(15)
obj.insertatbeg(20)
obj.insertatnode(18,20)
obj.delete(15)
obj.printlinklist()
````
