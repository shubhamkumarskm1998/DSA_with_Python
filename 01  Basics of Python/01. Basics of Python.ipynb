```python
def max_sum_rectangle_bruteforce(matrix):
    rows = len(matrix)
    cols = len(matrix[0])

    max_sum = float('-inf')

    best_top = best_bottom = 0
    best_left = best_right = 0

    for top in range(rows):

        for bottom in range(top, rows):

            for left in range(cols):

                for right in range(left, cols):

                    current_sum = 0

                    # Calculate rectangle sum
                    for r in range(top, bottom + 1):
                        for c in range(left, right + 1):
                            current_sum += matrix[r][c]

                    if current_sum > max_sum:
                        max_sum = current_sum

                        best_top = top
                        best_bottom = bottom
                        best_left = left
                        best_right = right

    return (
        max_sum,
        best_top,
        best_bottom,
        best_left,
        best_right
    )
```

```python
matrix = [
    [1, 2, -4],
    [8, 7, -1],
    [1, 1, -4]
]

result = max_sum_rectangle_bruteforce(matrix)

print(result)
```
