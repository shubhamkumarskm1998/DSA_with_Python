## Kadane's Algorithm

Kadane's Algorithm is a technique used to find the Maximum Sum Contiguous Subarray in an array in O(N) time.

### Brute Force

-- list = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

-- Maximum Sum = 6

-- Subarray = [4,-1,2,1]

```python
list = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
max_sum = float("-inf")  ## since negative numbers are there, if i choose max_sum =0 right then max of -1 and 0 will always be 0
start = end = 0
n = len(list)
for i in range(n):
    #print(f"starting index: {i}")
    cur_sum =0
    for j in range(i,n):
        #print(f"subarray: {list[i]} -> {list[j]}")
        cur_sum +=list[j]
        #print(f"sum: {cur_sum}")
        if cur_sum > max_sum:
            max_sum = cur_sum
            start = i
            end = j
        #print(f"max sum : {max_sum}")
print(f"max sum global: {max_sum}")
print(f"subarray: {list[start:end+1]}")
```

Worst Time Complexity = O(N*N)

Worst Space Complexity = O(1)

## Using Kadane's approach

```python
class kadane:
    def __init__(self,arr):
        self.arr = arr

    def max_sum(self):
        max_sum = curr_sum= self.arr[0]
        start=end=0
        temp_start=0

        for i in range(1, len(self.arr)):

            if self.arr[i] > curr_sum + self.arr[i]:
                curr_sum = self.arr[i]
                temp_start =i
            else:
                curr_sum += self.arr[i]

            if curr_sum > max_sum:
                max_sum = curr_sum
                max_sum = max(max_sum,curr_sum)
                start =temp_start
                end = i
        return max_sum, self.arr[start:end+1]

obj = kadane(list)
print(obj.max_sum())
```

O(N), O(1)

## Max Sum Rectangle (DP and Kadane's Algorithm)

## Brute Force

```python
matrix = [ [1,2,-4],
           [8,7,-1],
           [1,1,-4]]

max_sum = float("-inf")
for top in range(len(matrix)):
    #print(f"Top : {top}, {matrix[top]}")
    for bottom in range(top,len(matrix)):
        #print(f"Bottom : {bottom}, {matrix[top:bottom+1]}")
        for left in range(len(matrix[0])):
            #print(f"left : {left}, {matrix[top][left]}")
            for right in range(left, len(matrix[0])):
                #print(f"right : {right}, {matrix[bottom][right]}")
                curr_sum=0
                for r in range(top,bottom+1):
                    for c in range(left,right+1):
                        #print(f" row:{r},column:{c}")
                        curr_sum +=matrix[r][c]
                        #print(curr_sum)
                if curr_sum > max_sum:
                    max_sum = curr_sum
                    final_top= top
                    final_left = left
                    final_bottom = bottom
                    final_right = right
print(f"max global sum : {max_sum}")
print(f"rectange -> ({final_top,final_left}),({final_bottom,final_right})")
```

O(n^6), O(1)

```python
class kadane:
    def __init__(self, matrix):
        self.matrix = matrix
        self.rows = len(matrix)
        self.cols = len(matrix[0])

    def kadane(self, arr):
        current_sum = arr[0]
        max_sum = arr[0]
        start = end = 0
        temp_start = 0

        for i in range(1, len(arr)):

            if arr[i] > current_sum:
                current_sum = arr[i]
                temp_start = i
            else:
                current_sum += arr[i]

            if current_sum > max_sum:
                max_sum = current_sum
                start = temp_start
                end = i
        return max_sum, start, end

    def max_sum_rectangle(self):
        max_sum = float("-inf")
        final_top = final_bottom = 0
        final_left = final_bottom = 0

        for left in range(self.cols):
            temp = [0] * self.rows

            for right in range(left, self.cols):

                for row in range(self.rows):
                    temp[row] += self.matrix[row][right]

                current_sum, top, bottom = self.kadane(temp)

                if current_sum > max_sum:
                    max_sum = current_sum
                    final_left = left
                    final_right = right

                    final_top = top
                    final_bottom = bottom

        rect = []
        for i in range(final_top, final_bottom + 1):
            rect.append(self.matrix[i][final_left:final_right + 1])

        return {
            "max_sum": max_sum,
            "final_top": final_top,
            "final_bottom": final_bottom,
            "final_left": final_left,
            "final_right": final_right,
            "rect": rect}


matrix = [[1, 8, 1],
          [2, 7, 1],
          [-4, -1, -4]]

obj = kadane(matrix)

result = obj.max_sum_rectangle()

print(result)
```

O[n^3],O(1)
