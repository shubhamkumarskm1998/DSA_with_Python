This is the famous "Find Missing and Repeating Number" problem.

Given an array of size n containing numbers from 1 to n:

- One number appears twice
- One number is missing

Example:

arr = [1,2,2,4,5]

Repeating = 2

Missing   = 3

#### brute force approach

```python
arr = [1,2,2,4,5]
for i in range(1,len(arr)+1):
    count =0
    for j in arr:
        if i == j:
            count +=1
    if count == 0:
        print("missing:",i)
    if count ==2:
        print("repeating:",i)
```

### O(N) + O(N) = O(N^2)

### Hashmap approach

```python
arr = [1,2,2,4,5]
hash={}
count=0
for i in range(len(arr)):
    if arr[i] in hash:
        hash[arr[i]] +=1
    else:
        hash[arr[i]] =1
for i in range(1,len(arr)+1):
    if i not in hash:
        print("missing:",i)
    elif hash[i] ==2:
        print("repeating:",i)
```

O(N) and O(N)

### Approach 3: Math Equation : sum of continous n elements = n(n+1)/2

let x be repeating and y be missing, observations
- x-y = actual_sum - epxected_sum
- x^2-y^2 = actual_sq_sum - expected_sq_sum
- then calculate x+y = (actual_sq_sum - expected_sq_sum)/x-y
- solve for x and y
- x - y + x-y = difference + sum_sq //2

```python
arr = [1,2,2,4,5]
n = len(arr)
expected_sum = n*(n+1)//2
actual_sum = sum(arr)
difference = actual_sum - expected_sum

expected_sq_sum = n*(n+1)*(2*n+1)//6
actual_sq_sum = sum([i*i for i in arr])

sum_xy = (actual_sq_sum -expected_sq_sum )//difference

repeating = (sum_xy +difference)//2
missing = (sum_xy-difference)//2
print("repeating",repeating)
print("missing",missing)
```

### XOR Approach :

- X ^ X = 0
- X ^ Y = 1

- if we do xor all the elements will cancel out expect for 2 ^ 3, which is repeating and missing
- then we do set_bit comparision and then do comparison and again we will get repeating and missing

```python
arr = [1, 2, 2, 4, 5]
n = len(arr)
# Step 1: XOR array elements
array_xor = 0
for num in arr:
    array_xor ^= num
# Step 2: XOR numbers from 1 to n
expected_xor = 0
for i in range(1, n + 1):
    expected_xor ^= i
# Step 3: Missing ^ Repeating
missing_repeating_xor = array_xor ^ expected_xor
# Step 4: Find rightmost set bit
rightmost_set_bit = missing_repeating_xor & -missing_repeating_xor
# Step 5: Create two buckets
group_with_bit_set = 0
group_with_bit_not_set = 0

# Process array
for num in arr:
    if num & rightmost_set_bit:
        group_with_bit_set ^= num
    else:
        group_with_bit_not_set ^= num
# Process numbers 1 to n
for i in range(1, n + 1):
    if i & rightmost_set_bit:
        group_with_bit_set ^= i
    else:
        group_with_bit_not_set ^= i
print("Group 1 =", group_with_bit_set)
print("Group 2 =", group_with_bit_not_set)
# Determine which is repeating
if arr.count(group_with_bit_set) == 2:
    repeating_number = group_with_bit_set
    missing_number = group_with_bit_not_set
else:
    repeating_number = group_with_bit_not_set
    missing_number = group_with_bit_set

print("Repeating Number =", repeating_number)
print("Missing Number =", missing_number)
```

### New Question: Find the Duplicate Number problem using Floyd's Cycle Detection (Slow & Fast Pointer).

Array size = n + 1

Numbers range from:

1 to n

Only one number is repeated.

Example:

nums = [1,3,4,2,2]

Output:

2

```python
nums = [1,3,4,2,2]
slow,fast =0,0
while True:
    slow = nums[slow]
    fast = nums[nums[fast]]
    if slow == fast:
        break
slow=0
while slow != fast:
    slow = nums[slow]
    fast = nums[fast]
print(slow)
```
