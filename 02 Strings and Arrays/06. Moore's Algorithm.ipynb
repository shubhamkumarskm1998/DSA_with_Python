## Moore's Voting Algorithm (Majority Element)

#### Problem: Find the element that appears more than n/2 times in an array.

```python
nums = [2, 2, 1, 1, 1, 2, 2]
answer = 2
```

```python
## Bruteforce approach
nums = [2, 2, 1, 1, 1, 2, 2]
for i in range(len(nums)):
    count=0
    for j in range(len(nums)):
        if nums[i] == nums[j]:
            count +=1
    if count > len(nums)//2:
        print(f"Element {nums[i]} -> count: {count}")
        break
```

#### O(N^2) time and O(1) space

### using hashing- > hashmap

```python
nums = [2, 2, 1, 1, 1, 2, 2]
lookup = {}
for i in range(len(nums)):
    if nums[i] in lookup:
        lookup[nums[i]] +=1
    else:
        lookup[nums[i]] = 1
for key,value in lookup.items():
    if value > len(nums)//2:
        print(key)
        #break
```

## Moore' voting algorithm

- If a majority element exists (> n/2 times), it can never be completely canceled out by all other elements.

Keep:
- candidate
- count

- Rules
- If count == 0:
  -  candidate = current element

- If current element == candidate:
  -  count += 1
- Else:
  -  count -= 1

```python
nums = [2, 2, 1, 1, 1, 2, 2]
count=0
candidate=nums[0]
for i in range(len(nums)):
    if count==0:
        candidate = nums[i]

    if nums[i] == candidate:
        count = count + 1
    else:
        count -=1
print(candidate)
```

```python
nums = [2, 2, 1, 1, 1, 2, 2]
count=0
candidate = None
for i in range(len(nums)):
    print(f"{i}->{nums[i]}")
    if count ==0:
        candidate = nums[i]
        print(f"candidate -> {candidate}")
    if nums[i] == candidate:
        count += 1
        print(f"Count:{count} for candidate {candidate}")
    else:
        count -=1
        print(f"Count Reduced to { count }")
```
