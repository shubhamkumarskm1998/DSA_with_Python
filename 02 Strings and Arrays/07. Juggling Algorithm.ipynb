### Juggling Algorithm (Array Rotation)

The Juggling Algorithm is used to rotate an array by d positions efficiently.

## Example:

arr = [1, 2, 3, 4, 5, 6, 7]
d = 2

right rotate by 2:

[6,7,1,2,3,4,5]

```python
### Brute force approach -> Left Rotation
## Example:

arr = [1, 2, 3, 4, 5, 6, 7]
d = 2

for i in range(d): #number of rotations
    first=arr[0]
    for j in range(len(arr)-1): # of elements
        #print(f"Before - > {arr}")
        arr[j]=arr[j+1]
        #print(f"After - > {arr}")
    arr[-1] = first
    #print(f"After first index replacement - > {arr}")
print(arr)
```

```python
### Brute force approach -> Right Rotation
## Example:
#output = [6,7,1,2,3,4,5]
arr = [1, 2, 3, 4, 5, 6, 7]
d = 2

for i in range(1,d+1): #number of rotations
    last = arr[-1]
    for j in range(len(arr)-1,0,-1):
        arr[j]=arr[j-1]
    arr[0]=last
print(arr)
```

### Reversal Algorithm
arr = [1, 2, 3, 4, 5, 6, 7]
d = 2

- take arr[:d] - reverse this - [1,2] -> [2,1] --O(d)
- take arr[d:len(arr)] -- reverse this - [3,4,5,6,7]- > [7,6,5,4,3] --O(n-d)
- add them and reverse it again - [2,1,7,6,5,4,3] -> [3,4,5,6,7,1,2] - O(n)
- O(2N) and O(1)

```python
arr = [1, 2, 3, 4, 5, 6, 7]
d = 2

def reverse(arr,start,end):
    arr = arr[start:end+1]
    for i in range(len(arr)//2):
        arr[i],arr[len(arr)-i-1] = arr[len(arr)-i-1],arr[i]
    return arr
a=reverse(arr,0,d-1)
b=reverse(arr,d,len(arr)-1)
print(reverse(a+b,0,len(a)+len(b)))
```

## Juggling Algorithm
- arr = [1, 2, 3, 4, 5, 6, 7]
- d = 2
- GCD(len(arr),d) = (7,2) - > 1
- Iterations will be 2  and we will shfit every 2nd element by 2
- Iteration 1: 3 4 5 6 7 1 2 (started with 1st element and then took every 2nd element)

```python
arr = [1, 2, 3, 4, 5, 6, 7,8,9]
d = 3

def gcd(a,b):  ##O(min(a,b)
    result = 1
    for i in range(1,min(a,b)+1):
        if a%i==0 and b%i==0:
            result=i
    return result

gd = gcd(len(arr),d)
print(gd)
```

For interviews, the preferred GCD is Euclid's Algorithm: O(log(min(a,b)

- def gcd(a, b):

```python
def gcd(a,b):
    while b:
        a,b =b,a%b
    return a
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
## 4 2 3 7 5 6 1 8 9
## 4 5 3
d = 3
g = gcd(len(arr),d)

for i in range(g):
    temp = arr[i]
    j = i
    while True:
        k= (j+d)%len(arr)
        if k == i:
            break
        arr[j] = arr[k]
        j = k
    arr[j]=temp
print(arr)

```

```python
def gcd(a,b):
    while b:
        a,b =b,a%b
    return a
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
## 4 2 3 7 5 6 1 8 9
## 4 5 3
d = 3
g = gcd(len(arr),d)

for i in range(g):
    print(f"Iteration 1: {i}")
    temp = arr[i]
    print(f"temp: {i,arr[i]}")
    j= i
    while True:
        k =(j+d)%len(arr)
        print(f"k:{k}")
        if k == i:
            print(f"k==i")
            break
        arr[j] = arr[k]
        j=k
        print(f"arr before temp -> {arr}")
    arr[j] = temp
    print(f"arr after temp->{arr}")

print(arr)

```

1. Start from an index.
2. Save its value in temp.
3. Jump by d positions.
4. Copy the next value into the current position.
5. Continue jumping.
6. When you return to the start, put temp back.
7. Repeat for gcd(n,d) cycles.
