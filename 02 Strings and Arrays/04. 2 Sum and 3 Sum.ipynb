## 2 Sum Question

```python
## Brute force approach
nums = [2,11,7,15]
target = 18

for i in range(len(nums)):
    for j in range(i,len(nums)):
        if target == nums[i]+nums[j]:
            print([i,j])
            print(f"{nums[i]} + {nums[j]} --> {target}")
```

check every pair, O(N^2), O(1)

## Hashmap Approach
## 2nd approach will hashmap, where we check compliment = target -num and check if this available in our lookup

```python
nums =[2,11,7,15]
target =18
lookup ={}
for i in range(len(nums)):
    print(f"i:{i}, {nums[i]}")
    compliment = target - nums[i]
    print(f"Compliment: {compliment}")
    print(f"lookup:{lookup}")

    if compliment in lookup:
        print("find")
        print(f"i -> {i}, {nums[i]}, compliment ->{compliment}, compliment_index: {lookup[compliment]}")
        break
    else:
        lookup[nums[i]] = i
    print(lookup)
```

O(N), O(N) -> space complexity due to story of lookup value.

## Two Pointer Approach O(n) and O(1)
### This only works when the array is sorted.

```python
nums =[2,11,7,15]
target =18
start = 0
end =len(nums) - 1

for i in range(end):
    if nums[start]+nums[end] == target:
        print(start, end,nums[start],nums[end])
    elif nums[start]+nums[end] > target:
        end =end - 1
    else:
        start = start +1
```

```python
nums = [-1, 0, 1, 2, -1, -4]
target = 0
result =[]

for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        for k in range(j+1,len(nums)):
            if nums[i]+nums[j]+nums[k]==target:
                #print(f"{i},{j},{k} -> {nums[i]}+{nums[j]}+{nums[k]}")
                triplet = [nums[i],nums[j],nums[k]]
                #print(triplet)
                if triplet not in result:
                    result.append(triplet)
print(result)
```

## Using Hashset Approach
## fix one value nums[i] and find the the nums[j] + nums[k] such that it is equal to nums[j]+nums[k] = - nums[i]

```python
nums = [-1, 0, 1, 2, -1, -4]
target = -1
result =[]
for i in range(len(nums)):
    seen = set()
    two_sum = target -nums[i]
    for j in range(i+1, len(nums)):
        third = two_sum - nums[j]

        if third in seen:
            triplet = [nums[i],nums[j],third]
            result.append(triplet)
        seen.add(nums[j])
print(result)
```

O(N^2), O(N+N)=O(N)

### 3rd Approach -> # 3 Pointer (1 fixed and 2 Pointer) BEST O(N^2) and O(1)

```python
nums = [-1, 0, 1, 2, -1, -4]
target = -1
result = []
nums.sort()
for i in range(len(nums)-2):
    if i > 0 and nums[i] == nums[i-1]:
        continue
    left = i+1
    right = len(nums)-1

    while left <  right:
        total = nums[i]+nums[left]+nums[right]

        if total == target:
            result.append([nums[i],nums[left],nums[right]])
            left +=1
            right -=1

        elif total > target:
            right -=1
        else:
            left += 1

print(result)
```
