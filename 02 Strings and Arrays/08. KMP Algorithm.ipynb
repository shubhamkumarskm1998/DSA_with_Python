## Robin Karp Algorith --> Rolling Hash Method

Rabin-Karp uses hashing for pattern matching. First, we compute the hash of the pattern. Then we compute the hash of each text window of the same size. If the hashes match, we verify by comparing the actual strings because collisions can occur. To make it efficient, Rabin-Karp uses a rolling hash so the next window's hash can be calculated in O(1) time. Average complexity is O(N+M).

### Steps
- Compute hash of pattern.
- Compute hash of first window.
- Compare hashes.
- If hashes match, verify strings.
- Roll the hash to the next window.

```python
def hash_function(str):
    h = 0
    for i in str:
        h = h*2 + ord(i)    #Taking Base 2
    return h

text = "ababcabcabababd"
pattern = "ababd"

n = len(text)
m = len(pattern)
p_hash = hash_function(pattern)
w_hash = hash_function(text[:m])
highest_power=1
for i in range(m-1):
    highest_power *=2

for i in range(n-m+1):
    if p_hash == w_hash:
        if text[i:i+m] == pattern:
            print(f"found index at {i}")
    if i < n-m:
        #shift the text by one character
        w_hash -= highest_power*ord(text[i])
        w_hash *= 2
        w_hash +=ord(text[i+m])
```

### KMP Algorithm

The biggest question is:

Why do we need KMP when brute force already works?

Because brute force rechecks characters that we already know matched.

When mismatch occurs:

ABABD
ABABA
     ↑ mismatch

Ask:

"Inside the matched part ABAB, is there some prefix that is also a suffix?"

If yes, don't start from zero.

Reuse previous work.

## To use this work we create LPS array ( Longest Prefix suffix array)

```python
def lps(pattern):
    n = len(pattern)
    lps = [0]*n
    i=1
    length = 0
    while i < n:
        if pattern[i]==pattern[length]:
            length +=1
            lps[i] = length
            i+=1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1
    return lps

text = "ababcabcabababd"
pattern = "ababd"
lps = lps(pattern)
i = 0
j = 0
print(lps)
while i < len(text):
    print(f"Iteration {i}")
    if text[i] == pattern[j]:
        print(f"Increasing {i} and {j} : {text[i],pattern[j]}")
        i += 1
        j += 1
    if j == len(pattern):
        print("Found at", i-j)
        break
    elif i < len(text) and text[i] != pattern[j]:
        print(f"Different {i} and {j} and {len(text)} : {text[i],pattern[j]}")
        if j != 0:
            print(f"j is not zero and j is {j}")
            j = lps[j-1]
            print(f"j is changed through LPS {lps} j is > {j}")
        else:
            print(f"j is zero {j}")
            i += 1
```
