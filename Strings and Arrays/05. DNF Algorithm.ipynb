### DNF Algorithm
The Dutch National Flag Algorithm is a famous algorithm by Edsger W. Dijkstra used to sort an array containing only three distinct values (commonly 0, 1, and 2) in O(n) time and O(1) space.

nums = [2, 0, 2, 1, 1, 0]


#### We can sort this array right? using our sorting algo's which takes O(nlogn) and takes O(N) but we want a more optimized approach.

#### we can use bruteforce

```python
nums = [2, 0, 2, 1, 1, 0]
count_zero=0
count_one=0
count_two=0
for i in range(len(nums)):
    if nums[i] == 0:
        count_zero +=1
    elif nums[i] == 1:
        count_one +=1
    else:
        count_two +=1
nums =  [0]*count_zero + [1]*count_one + [2]*count_two
print(nums)
```

## here we have O(n) and O(1) as well but here we are counting frequencies and recontructing the array.

### DNF Algorithm

```python
nums = [2, 0, 2, 1, 1, 0]
left = mid =0
high = len(nums)-1
while mid <= high:
    if nums[mid]==0:
        nums[left],nums[mid] = nums[mid],nums[left]
        left +=1
        mid +=1
    elif nums[mid] == 1:
        mid += 1
    else:
        nums[mid],nums[high]=nums[high],nums[mid]
        high -=1
print(nums)
```
